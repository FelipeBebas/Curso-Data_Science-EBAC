#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
import pandas as pd
import yfinance as yf
```

```python
# Definindo os ativos do Yahoo Finance no período de 2 anos até a data atual
ativos = ['^BVSP', 'PETR4.SA', 'VALE3.SA']
dt_ini, dt_fim = '2023-01-01', '2025-07-01'
```

```python
# Baixando os dados
df_bvsp = yf.download(ativos, start=dt_ini, end=dt_fim)
```

```python
print("Verificando valores nulos:\n")
print(df_bvsp.isnull().sum())
```

```python
# Exportando df_bvsp para um CSV para download
df_bvsp.to_csv('df_bvsp.csv')
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Criando uma coluna de amplitude para cada ativo
df_bvsp['Amplitude', 'PETR4.SA'] = df_bvsp['High', 'PETR4.SA'] - df_bvsp['Low', 'PETR4.SA']
df_bvsp['Amplitude', 'VALE3.SA'] = df_bvsp['High', 'VALE3.SA'] - df_bvsp['Low', 'VALE3.SA']
df_bvsp['Amplitude', '^BVSP'] = df_bvsp['High', '^BVSP'] - df_bvsp['Low', '^BVSP']
display(df_bvsp.head())
```

```python
import plotly.graph_objects as go
import plotly.express as px

# Seleciona apenas as colunas de Amplitude do DataFrame df_bvsp.
df_amplitude = df_bvsp['Amplitude'].copy()

# Isso ajuda o Plotly Express a identificar qual coluna usar para colorir as linhas.
df_amplitude.columns.name = 'Ativo'

# O melt() serve para transformar o DataFrame de largo para longo.
df_amplitude = df_amplitude.reset_index().melt(id_vars='Date', var_name='Ativo', value_name='Amplitude Diária')

# Cria o gráfico de linhas usando plotly
fig_amplitude = px.line(df_amplitude,
                        x='Date',
                        y='Amplitude Diária',
                        color='Ativo', # Diferencia as linhas por ativo
                        title='Amplitude Diária (Máxima - Mínima) dos Ativos ao Longo do Tempo')

# Personaliza o layout
fig_amplitude.update_layout(
    xaxis_title='Data',
    yaxis_title='Amplitude (R$)',
    legend_title='Ativo',
    hovermode='x unified' # Agrupa as informações dos ativos na mesma data ao passar o mouse
)

# Exibe o gráfico.
fig_amplitude.show()
```

O índice BVSP apresentou maior volatilidade em comparação aos outros dois.
